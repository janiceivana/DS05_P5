# HOBBIES_1 opti price

```python
hobbies_1=my_data[my_data['dept_id']=="HOBBIES_1"]
hobbies_1 = hobbies_1.drop(hobbies_1[hobbies_1['unit_sale'] == 0].index)
hobbies_1.head()
```

```python
h1_df = pd.DataFrame(columns=['item_id', 'dept_id', 'cat_id', 'state_id', 'price',  'sale_perday', 'price_change', 'sale_change', 'elasticity'])

# Filter the DataFrame once
filtered_df = hobbies_1.groupby(['item_id', 'dept_id', 'cat_id', 'state_id']).filter(lambda x: len(x) > 0)

for (item_id, dept_id, cat_id, state_id), group_df in filtered_df.groupby(['item_id', 'dept_id', 'cat_id', 'state_id']):
    for sell_price in group_df['sell_price'].unique():
        item_df = group_df[group_df['sell_price'] == sell_price]
        sale_perday = item_df['unit_sale'].mean()

        hobbies_1_df = pd.DataFrame({'item_id': [item_id], 'dept_id': [dept_id], 'cat_id': [cat_id], 'state_id': [state_id], 'price': [sell_price],'sale_perday': [sale_perday]})
        hobbies_1_df = hobbies_1_df.dropna()

        h1_df = h1_df.append(hobbies_1_df, ignore_index=True)

h1_df['price_change'] = h1_df.groupby(['item_id', 'dept_id', 'cat_id', 'state_id'])['price'].pct_change() 
h1_df['sale_change'] = h1_df.groupby(['item_id', 'dept_id', 'cat_id', 'state_id'])['sale_perday'].pct_change() 
h1_df['elasticity'] = -h1_df['sale_change'] / h1_df['price_change']

h1_df = h1_df.dropna()
```

```python
h1_df['elasticity'] = h1_df['elasticity'].apply(lambda x: 0 if x < 0 else x)
h1_df
```

```python
h1_df["target_days"] = 7

grouped = h1_df.groupby(['item_id', 'state_id'])
h1_df['cost_price'] = grouped['price'].transform(lambda x: 0.5 * x.min())

discounts = [0, 10, 30,50]
new_rows = []
for index, row in h1_df.iterrows():
    for discount in discounts:
        new_row = row.copy()
        new_row['discount%'] = discount
        new_rows.append(new_row)

h1_opti = pd.DataFrame(new_rows)
h1_opti = h1_opti[["item_id","state_id","price","sale_perday","elasticity","target_days","discount%","cost_price"]]

h1_opti["stock on hand"] = 100

h1_opti["discount_price"] = (1-h1_opti["discount%"]/100)*h1_opti["price"]

h1_opti["predict_total_sale"] = (h1_opti["target_days"]*h1_opti["sale_perday"])*(1+abs((h1_opti["discount_price"]-h1_opti["price"])/h1_opti["price"])*h1_opti["elasticity"])
h1_opti.reset_index(drop=True, inplace=True)
h1_opti.loc[h1_opti["predict_total_sale"] > h1_opti["stock on hand"], "predict_total_sale"] = h1_opti["stock on hand"]

h1_opti
```

```python
h1_opti["cost"] = round(h1_opti['cost_price']*h1_opti["predict_total_sale"],2)

h1_opti["revenue"] = round(h1_opti["discount_price"]*h1_opti["predict_total_sale"],2)

h1_opti["profit"] = h1_opti["revenue"]-h1_opti["cost"]

h1_opti
```

```python
h1_opti = h1_opti.loc[h1_opti.groupby(['item_id', 'state_id','price'])['profit'].idxmax()]
h1_opti.reset_index(drop=True, inplace=True)
h1_opti
```
